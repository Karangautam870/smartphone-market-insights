```python
def plot_graph(col_name):
    sns.histplot(df[col_name], bins=30, kde=True,label=col_name)
    sns.catplot(kind ='box', data=df,x=col_name)
    
```

```python
# using knn imputer to fill missing values in rating column
from sklearn.impute import KNNImputer

x_df = df.select_dtypes(include=['float64', 'int64']).drop(columns=['price'])

imputer = KNNImputer(n_neighbors=5)
```

```python
x_df_val = imputer.fit_transform(x_df)
```

```python
x_df = pd.DataFrame(x_df_val, columns=x_df.columns)
```

```python
x_df['price'] = df['price']
```

```python
x_df.head()
```

```python
a = x_df.corr(numeric_only=True)['price'].reset_index()
```

```python
b =  df.corr(numeric_only=True)['price'].reset_index()
```
